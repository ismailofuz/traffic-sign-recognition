```python
import os
from PIL import Image
import numpy as np

data = []
labels = []
classes = 43  # Sifatlar soni

# Rasm va etiketlarni olish
for i in range(classes):
    path = os.path.join(cur_path, 'Train', str(i))  # To‘g‘ri path
    images = os.listdir(path)

    for a in images:
        image_path = os.path.join(path, a)  # To‘g‘ri rasm yo‘lini olish
        try:
            image = Image.open(image_path)  # Rasmni ochish
            image = image.resize((30, 30))  # Rasm o‘lchamini o‘zgartirish
            image = np.array(image)  # Rasmni numpy arrayga o‘zgartirish

            data.append(image)  # Rasmni qo‘shish
            labels.append(i)  # Tegni (class) qo‘shish
        except Exception as e:
            print(f"Error loading image {a}: {e}")

print(f"Loaded {len(data)} images.")
print(f"Loaded {len(labels)} labels.")
```

```python
print(len(data))
```

```python
x = data # data contains the image arrays
y = labels # labels contains the corresponding class labels
```

```python
#Converting lists into numpy arrays
x = np.array(x)
y = np.array(y)
```

```python
X_train, X_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train = X_train/255.0
X_val = X_val/255.0
```

```python
print("Shape of train images is:", X_train.shape)
print("Shape of labels is:", y_train.shape)
```

```python
#Building the model
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(30, 30, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())  # this converts our 3D feature maps to 1D feature vectors
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(43))
model.add(Activation('softmax'))

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
aug = ImageDataGenerator(
    rotation_range=10,
    zoom_range=0.15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.15,
    horizontal_flip=False,
    vertical_flip=False,
    fill_mode="nearest")

model.fit(aug.flow(X_train, y_train, batch_size=32), epochs=15, validation_data=(X_val, y_val))

model.save('/content/drive/MyDrive/Colab Notebooks/Neural network/German Traffic Sign Recognition Benchmark/MODEL_TRAFFIC_SIGN_RECOGNITION.h5')
  # always save your weights after training or during training
```
